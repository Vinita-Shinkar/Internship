# Q2. Write a python program to scrape data for "Data Scientist" Job position in "Bangalore" location. You have to scrape the job-title, job-location, company-name, full job-description. You have to scrape first 10 jobs data. Website - https://www.naukri.com/

```python
# Importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
```

```python
# Let's first connect with the web driver
driver = webdriver.Chrome(r"C:\Users\vinit\Downloads\chromedriver.exe")
```

```python
url_1 = "https://naukri.com/"
driver.get(url_1)
```

```python
# Finding element for job search using xpath
search_job = driver.find_element_by_xpath("//input[@name='keyword']")

# Write on search bar
search_job.send_keys("Data Scientist")
```

```python
# Finding element for job location bar using xpath
search_loc = driver.find_element_by_xpath("//input[@name='location']")

# Write on location bar 
search_loc.send_keys("Bangalore")
```

```python
# Finding element for search button using xpath
search_btn = driver.find_element_by_xpath("//button[@class='btn']")

# Click the search button
search_btn.click()
```

```python
# Finding new url for the job titles using xpath
job_url = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")

Job_Url = []
for i in job_url:
    Job_Url.append(i.get_attribute("href"))

Job_Titles = []
for i in Job_Url:
    driver.get(i)
    # Extracting Job Title using xpath
    try:
        job_titles = driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        Job_Titles.append(job_titles.text)
    except:
        Job_Titles.append("--")
        
Job_Locations = []
for i in Job_Url:
    driver.get(i)
    # Extracting Job Location using xpath
    try:
        job_locations = driver.find_element_by_xpath("//div[@class='loc']//span//a")
        Job_Locations.append(job_locations.text)
    except:
        Job_Locations.append("--")
        
Company_Name = []
for i in Job_Url:
    driver.get(i)
    # Extracting Company Name using xpath
    try:
        company_name = driver.find_element_by_xpath("//div[@class='jd-top-head']//div//a[1]")
        Company_Name.append(company_name.text)
    except:
        Company_Name.append("--")

Job_Desc = []
for i in Job_Url:
    driver.get(i)
    # Extracting Job Description using xpath
    try:
        job_desc = driver.find_element_by_xpath("//div[@class='dang-inner-html']")
        Job_Desc.append(job_desc.text.replace("\n",""))
    except:
        Job_Desc.append("--")  
```

```python
# import pandas as pd
ques_2 = pd.DataFrame({})
ques_2["JOB TITLE"] = Job_Titles[1:17]
ques_2["JOB LOCATION"] = Job_Locations[1:17]
ques_2["COMPANY NAME"] = Company_Name[1:17]
ques_2["FULL JOB DESCRIPTION"] = Job_Desc[1:17]
ques_2.index+=1

# DataFrame of 13 Jobs Data
ques_2
```
