# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release)

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
pd.options.display.max_rows = 100
```

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in soup.select('td.ratingColumn strong')]

imdb = []
```

```python
for index in range(0, len(movies)):
    # Seperate movie into: 'place', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    imdb.append(data)

  
```

```python
imdb_dataframe = pd.DataFrame(imdb)
imdb_dataframe.sort_values(by=['rating'], ascending=False)
imdb_dataframe.head(100)
```

# Q9. Write a python program to scrape fresher job listings from 'https://internshala.com/'. It should include job title, company name, CTC and apply date. 

```python
from bs4 import BeautifulSoup
import requests
jobs = requests.get("https://internshala.com/fresher-jobs")
jobs
```

```python
soup_12 = BeautifulSoup(jobs.content)

job_title = soup_12.find_all("div", class_="heading_4_5 profile")
Job_Title = []
for i in job_title:
    Job_Title.append(i.text.replace("\n",""))

company_name = soup_12.find_all("div", class_="heading_6 company_name")
Company_Name = []
for i in company_name:
    Company_Name.append(i.text.replace("\n","").strip())

ctc = soup_12.find_all("div", class_="item_body")
CTC = []
for i in ctc:  
    CTC.append(i.text)

CTC_New = []
for i in CTC:
    if 'LPA' in i:
        CTC_New.append(i.strip())

applydate = soup_12.find_all("div", class_="item_body")
ApplyDate=[]
for i in range(2,len(applydate),3):
    x = (str(applydate[i]).split('>'))
    ApplyDate.append(x[1].split('<')[0])

import pandas as pd
Fresher_Job = pd.DataFrame({})
Fresher_Job['JOB TITLE'] = Job_Title
Fresher_Job['COMPANY NAME'] = Company_Name
Fresher_Job['CTC'] = CTC_New
Fresher_Job['APPLY DATE'] = ApplyDate
Fresher_Job.index+=1
Fresher_Job
```
